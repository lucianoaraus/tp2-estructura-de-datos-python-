# **TDA Pila:**

```python


# TDA
class Pila:
  def __init__(self,entrada = None):                         # Constructor
    self.pila = []
    if entrada:
      for dato in entrada:
        self.pila.append(dato)
  
  def __repr__(self):
    return str(self.pila)

  def vaciar(self):                                          # Vacia la pila
    self.pila.clear()

  def estaVacia(self):
    return len(self.pila) == 0                               # Pila vacia

  def apilar(self, item):                                    # Suma un item a la pila
    self.pila.append(item)

  def desapilar(self):                                       # Retira el ultimo elemento de la pila (si hay)
    dato = None
    if not self.estaVacia():
      dato = self.pila.pop()
    return dato

  def primerElemento(self):                                  # Devuelve el primer elemento de la pila (si hay)
    if self.estaVacia():
      print("La lista se encuentra vacia.")
      return None
    return self.pila[-1]

  def clonar(self):                                         # Devuelve una copia de la pila
    pilaClon = Pila()
    for elemento in self.pila:
      pilaClon.pila.append(elemento) #### pilaClon.apilar(elemento)
    return pilaClon

  def tamano(self):                                          # Devuelve el tamaño de la pila
    #print("El tamaño de la pila es:")
    return len(self.pila)

  def estaVacia(self):                                       # Devuelve TRUE si la pila se encuentra vacia
    return len(self.pila) == 0

  # Funciones Auxiliares
  def invertirPila(self):
    pilaAux = self.clonar()
    self.vaciar()
    while not pilaAux.estaVacia():
      self.apilar(pilaAux.desapilar())

```

```python
from enum import Enum

class GeneroLibro(int,Enum):
  Teatro = 0
  Poesia = 1
  Narracion = 2

#Se encarga de validar si el codigo de identificacion ingresado
#es valido para un libro
def validaCodigoDeIdentificacion(ID):
  if not (len(ID) == 8 and ID[0:3].isalpha() and ID[3:].isnumeric()):
    raise Exception ("ID incorrecto")
  else:
    return ID

#Se encarga de validar si el genero ingresado 
#pertenece a los generos existentes 
def validaGenero(genero):
  if not isinstance(GeneroLibro(genero) , GeneroLibro):
      raise Exception("Genero incorrecto")
  else:
    return genero



# TDA
class Libro:
  #constructor del TDA Libro, que recibe como parametro un codigo de identificacion,
  #un genero y una nacionalidad
  def __init__(self,codigoDeIdentificacion,genero,nacionalidad):
    self.codigoDeIdentificacion = validaCodigoDeIdentificacion(codigoDeIdentificacion)
    self.generos = validaGenero(genero)
    self.nacionalidad = nacionalidad
  
  #se encarga de definir como deberia mostrarse el TDA a la hora de ser imprimido
  def __repr__(self):                                        
    return "Cod_Identificacion:" + " " + str(self.codigoDeIdentificacion) + ", " + "Genero:" + " " + str(GeneroLibro(self.generos).name) + ", " + "Nacionalidad:" + " " + str(self.nacionalidad)



#Pruebas
TipoDeGeneros = GeneroLibro.Teatro.name
print(TipoDeGeneros)
libro1 = Libro("aaa99999",0,"Nacional")
print(libro1)
libro2 = Libro("aaa99999",1,"Nacional")
print(libro2)
 
```

```python
# Validaciones
def validaNumID(idDeEstanteria):
  if 0 < idDeEstanteria < 999:
    return idDeEstanteria
  else:
    raise Exception("El ID ingresado excede los limites.")


# TDA
class Estanteria: 
  # Constructor
  def __init__(self,numID = int,cantCritica = 50):
    self.pilaDeAutoresNacionales = Pila()
    self.pilaDeAutoresInternacionales = Pila()   
    self.numID = validaNumID(numID)
    self.cantCritica = cantCritica # Por defecto 50
                          
  def __repr__(self):
    return "Numero de ID:" + " " + str(self.numID) + str(self.pilaDeAutoresNacionales) + str(self.pilaDeAutoresInternacionales)

  #esta funcion se encarga de guardar el libro que reciba por parametro
  #en una de las dos pilas del TDA dependiendo de la nacionalidad de libro
  def guardarLibro(self,libro):
    if libro.nacionalidad == "Nacional" and self.pilaDeAutoresNacionales.tamano() < self.cantCritica: 
      self.pilaDeAutoresNacionales.apilar(libro)
    elif libro.nacionalidad == "Nacional" and self.pilaDeAutoresNacionales.tamano() >= self.cantCritica: 
      self.pilaDeAutoresNacionales.apilar(libro)
      #raise Exception("Se ha alcanzado la cantidad critica en la estanteria.")

    if libro.nacionalidad == "Internacional" and self.pilaDeAutoresInternacionales.tamano() < self.cantCritica:
      self.pilaDeAutoresInternacionales.apilar(libro)
    elif libro.nacionalidad == "Internacional" and self.pilaDeAutoresInternacionales.tamano() >= self.cantCritica:
      self.pilaDeAutoresInternacionales.apilar(libro)
      #raise Exception("Se ha alcanzado la cantidad critica en la estanteria.")
  
  #guardarLibro(self,libro): Se definen 2 bloques segun la nacionalidad del libro (Nacionales e Internacionales).
  #En primera instancia, se hace una comparacion con su nacionalidad. Seguido de eso, se chequea el tamaño de la pila y, si es inferior a la cantidad critica establecida (50) lo apila en la pila de libros nacionales de la estanteria, y sino, lo apila y a su vez informa que la cantidad critica fue superada.
  #En caso de que la nacionalidad  del libro pasado por parametro sea la segunda opcion, se lleva a cabo el mismo procedimiento, apilando el libro en la pila de libros correspondiente.
  

  #esta funcion retorna las 2 pilas de autores que contiene el estante
  def pilasDeAutores(self):
    return self.pilaDeAutoresNacionales, self.pilaDeAutoresInternacionales
  
  
  #esta funcion se encarga de retornar el primer libro disponible de la estanteria, partiendo de los libros nacionales. Si las 2 pilas se encuentran vacias, se lanza una excepcion informandolo.
  def primerLibroDisponible(self):
    primerElemento = None
    if not self.pilaDeAutoresNacionales.estaVacia():
      primerElemento = self.pilaDeAutoresNacionales.primerElemento()
    elif not self.pilaDeAutoresInternacionales.estaVacia():
       primerElemento = self.pilaDeAutoresInternacionales.primerElemento()
    else:
      primerElemento = "La estanteria esta vacia."
    return primerElemento  

  #esta funcion se encarga de devolver el primer libro que encuentre con el genero que se pase por parametro,
  #primero busca en la pila de autores nacionales y luego en la pila de autores internacionales
  def libroParaRecomendar(self,generoDeLibro):
    libroARecomendar = None  
    
    libroARecomendar = self.recomendarLibroNacional(generoDeLibro)
    
    if libroARecomendar == None:
      libroARecomendar = self.recomendarLibroInternacional(generoDeLibro)

    if libroARecomendar == None:
      raise Exception("No hay un libro que se pueda recomendar en estos momentos.")
    
    return libroARecomendar      

  #Recomienda un libro de autor nacional del genero pasado por parametro. Clona la pila de autores nacionales y la guarda en una variable.
  #Mientras no este vacia, y si el genero del primer libro de la pila es igual al genero pasado por parametro, guarda el libro en una variable
  #que luego retornara. El libro a su vez es removido de la pila auxiliar.
  def recomendarLibroNacional(self,generoDeLibro):
    libroARecomendar = None
    auxNacional = self.pilaDeAutoresNacionales.clonar()
    if not auxNacional.estaVacia() : #Redundante
      while not auxNacional.estaVacia() :
        primerElemento = auxNacional.primerElemento()
        if primerElemento.generos == generoDeLibro:
          libroARecomendar = auxNacional.primerElemento()
        auxNacional.desapilar()
    return libroARecomendar    

  #Recomienda un libro de autor internacional del genero pasado por parametro. Clona la pila de autores internacionales y la guarda en una variable.
  #Mientras no este vacia, y si el genero del primer libro de la pila es igual al genero pasado por parametro, guarda el libro en una variable
  #que luego retornara. El libro a su vez es removido de la pila auxiliar.
  def recomendarLibroInternacional(self,generoDeLibro):
    libroARecomendar = None
    auxInternacional = self.pilaDeAutoresInternacionales.clonar()

    if not auxInternacional.estaVacia():
      while not auxInternacional.estaVacia():
        primerElemento = auxInternacional.primerElemento()
        if primerElemento.generos == generoDeLibro:
          libroARecomendar = auxInternacional.primerElemento()
        auxInternacional.desapilar()
    return libroARecomendar

  #esta funcion se encarga de buscar un libro y devolverlo, a partir del codigo del libro
  #pasado por parametro
  def buscarLibro(self,codigoLibro):
    libroBuscado = None
    libroBuscado = self.buscarLibroNacional(codigoLibro)
    if libroBuscado == None:
      libroBuscado = self.buscarLibroInternacional(codigoLibro)
    if libroBuscado == None:
      raise Exception("no se encontro el libro")
    return libroBuscado

  #esta funcion se encarga de retornar un libro, el cual se busca mediante el
  #codigo pasado por parametro, y lo remueve de la pila donde se encuentre
  def prestarLibro(self,codigoLibro):  
    libroParaMostrar = None
    
    libroParaMostrar = self.prestarLibroNacional(codigoLibro)

    if libroParaMostrar == None:

      libroParaMostrar = self.prestarLibroInternacional(codigoLibro)
    
    if libroParaMostrar == None:
      raise Exception("no hay un libro para prestar del genero deseado")
    return libroParaMostrar
  
  #esta funcion se encarga de devolver el mañano de las pilas de autores nacionales e internacionales 
  def librosPorTipo(self):
    return self.tamanoDePilaDeAutoresNacionales(),self.tamanoDePilaDeAutoresInternacionales()
  
  #esta funcion se encarga de devolver la cantidad total de libros que tiene el estante
  def cantidadTotalLibros(self):
    return self.tamanoDePilaDeAutoresNacionales() + self.tamanoDePilaDeAutoresInternacionales()

  #esta funcion se encarga de saber si el estante es critico, osea si sus estanterias superan la cantidad de 50 libros
  def esCritica(self):
    return self.tamanoDePilaDeAutoresNacionales() > 50 or self.tamanoDePilaDeAutoresInternacionales() > self.cantCritica # Se reemplazo el tamaño critico por 2 para simplificar procesos

  #esta funcion se encarga de retornar una pila con los libros en el estante
  #que pertenezcan al genero pasado por parametro  
  def librosPorGenero(self,generoLibro):
    totalLibrosPorGenero = 0
    totalLibrosPorGenero += self.librosNacionalesPorGenero(generoLibro) +  self.librosInternacionalesPorGenero(generoLibro)
    return totalLibrosPorGenero

  
  ###Funciones auxiliares###
  
  #esta funcion se encarga de retornar el tamanio de la pila
  #de autores nacionales
  def tamanoDePilaDeAutoresNacionales(self):
    return self.pilaDeAutoresNacionales.tamano()

  #esta funcion se encarga de retornar el tamanio de la pila
  #de autores internacionales
  def tamanoDePilaDeAutoresInternacionales(self):
    return self.pilaDeAutoresInternacionales.tamano()

  #esta funcion se encarga de prestar un libro de la pila de autores nacionales
  #del genero pasado por parametro
  def prestarLibroNacional(self,codigoLibro):
    libroParaMostrar = None
    nacionalesAux = self.pilaDeAutoresNacionales.clonar()
    nacionalesAux.invertirPila()
    self.pilaDeAutoresNacionales.vaciar()
    
    while not nacionalesAux.estaVacia():
      if nacionalesAux.primerElemento().codigoDeIdentificacion != codigoLibro:
        self.pilaDeAutoresNacionales.apilar(nacionalesAux.desapilar())
      else:
        libroParaMostrar = nacionalesAux.desapilar()
        while not nacionalesAux.estaVacia():
          self.pilaDeAutoresNacionales.apilar(nacionalesAux.desapilar())
    return libroParaMostrar

  #esta funcion se encarga de prestar un libro de la pila de autores internacionales
  #del genero pasado por parametro
  def prestarLibroInternacional(self,codigoLibro):
    libroParaMostrar = None
    internacionalesAux = self.pilaDeAutoresInternacionales.clonar()
    internacionalesAux.invertirPila()
    self.pilaDeAutoresInternacionales.vaciar()
    
    while not internacionalesAux.estaVacia():
      if internacionalesAux.primerElemento().codigoDeIdentificacion != codigoLibro:
        self.pilaDeAutoresInternacionales.apilar(internacionalesAux.desapilar())
      else:
        libroParaMostrar = internacionalesAux.desapilar()
        while not internacionalesAux.estaVacia():
          self.pilaDeAutoresInternacionales.apilar(internacionalesAux.desapilar())
    return libroParaMostrar

  #esta funcion se encarga de buscar un libro de autor nacional, y lo devuelve. Para llevarlo a cabo, se clona la pila de autores nacionales y se guarda como una pila auxiliar.
  #mientras la pila auxiliar no este vacia y el codigo del libro que se vaya a desapilar sea el mismo al codigo buscado, se guarda el libro que se desapilo (para luego ser retornado) y se termina de vaciar la lista (si tuviera libros).
  #en el caso de que no la pila este vacia o no se encuentre el codigo de libro buscado, se devuelve None.
  def buscarLibroNacional(self,codigoLibro):
    libroBuscado = None
    nacionalesAux = self.pilaDeAutoresNacionales.clonar()
    while not nacionalesAux.estaVacia():
      libro = nacionalesAux.desapilar()
      if libro.codigoDeIdentificacion == codigoLibro:
          libroBuscado = libro
    return libroBuscado  
  
  #esta funcion se encarga de buscar un libro de autor internacional, y lo devuelve. Para llevarlo a cabo, se clona la pila de autores internacionales y se guarda como una pila auxiliar.
  #mientras la pila auxiliar no este vacia y el codigo del libro que se vaya a desapilar sea el mismo al codigo buscado, se guarda el libro que se desapilo (para luego ser retornado) y se termina de vaciar la lista (si tuviera libros).
  #en el caso de que no la pila este vacia o no se encuentre el codigo de libro buscado, se devuelve None.
  def buscarLibroInternacional(self,codigoLibro):
    libroBuscado = None
    internacionalesAux = self.pilaDeAutoresInternacionales.clonar()
    while not internacionalesAux.estaVacia():
        libro = internacionalesAux.desapilar()
        if libro.codigoDeIdentificacion == codigoLibro:
          libroBuscado = libro
    return libroBuscado    
  
  
  #esta funcion se encarga de retornar la cantidad de libros nacionales
  #pertenecientes al genero pasado por parametro
  def librosNacionalesPorGenero(self,generoLibro):
    totalLibrosPorGenero = 0
    nacionalesAux = self.pilaDeAutoresNacionales.clonar()
    nacionalesAux.invertirPila()
    self.pilaDeAutoresNacionales.vaciar()

    while not nacionalesAux.estaVacia():
      elemento = nacionalesAux.desapilar()
      if elemento.generos != generoLibro:
        self.pilaDeAutoresNacionales.apilar(elemento)
      else:
        totalLibrosPorGenero += 1
        self.pilaDeAutoresNacionales.apilar(elemento)
     
    return totalLibrosPorGenero

  #esta funcion se encarga de retornar la cantidad de libros internacionales
  #pertenecientes al genero pasado por parametro
  def librosInternacionalesPorGenero(self,generoLibro):
    totalLibrosPorGenero = 0
    internacionalesAux = self.pilaDeAutoresInternacionales.clonar()
    internacionalesAux.invertirPila()
    self.pilaDeAutoresInternacionales.vaciar()
    
    while not internacionalesAux.estaVacia():
      elemento = internacionalesAux.desapilar()
      if elemento.generos != generoLibro:
        self.pilaDeAutoresInternacionales.apilar(elemento)
      else:
        totalLibrosPorGenero += 1
        self.pilaDeAutoresInternacionales.apilar(elemento)

    return totalLibrosPorGenero 
    
```
